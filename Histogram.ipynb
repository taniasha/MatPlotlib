```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Age': [18, 20, 22, 22, 23, 25, 30, 35, 40, 45]
}

df = pd.DataFrame(data)

plt.hist(df['Age'])
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

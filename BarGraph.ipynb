```python
#Bar Chart - used to compare categorical values using recatangul0ar bars
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Example 1
children = [10,20,40,10,50]
colors = ['red', 'blue','green','yellow','pink']

plt.bar(colors, children, color='green')
plt.xlabel('colors')
plt.ylabel('children')
plt.title('Colors-children')
```

```python
#horizontal bar chart
plt.barh(colors, children, color='green')
```

```python
#example with a dataset
batsman = {
    "batsman" : ['AB de Villiers', 'DA Warner','MS Dhoni','RG Sharma', 
                 'V Kohli'],
     '2015':[513, 562, 372, 482, 505],
     '2016':[687, 848, 284, 489, 973],
     '2017':[216, 641, 290, 333, 308]
}
df = pd.DataFrame(batsman)
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
plt.bar(np.arange(df.shape[0])-0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0])+0.2, df['2017'], width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])

```

```python
#Problem 
children = [10,20,40,10,50]
colors = ['red red red red red', 'blue blue blue blue blue','green green greeen green','yellow yellow yellow yellow yellow ','pink pink pink pink']

plt.bar(colors, children, color='orange')
plt.xticks(rotation='vertical') #otherwise the x axis label do overlap
```

```python
#Stacked Bar Chart
plt.bar(df['batsman'],df['2017'], label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016']+df['2017']),label='2015')

plt.legend()
plt.show()
```

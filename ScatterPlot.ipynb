```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    'Hours_studies':[1,2,3,4,5,6],
    'Marks':[35,40,43,49,55,90]
}
df = pd.DataFrame(data)
df
```

```python
plt.scatter(df['Hours_studies'], df['Marks'], marker='+', color='brown') 
plt.xlabel('Hours_studies')
plt.ylabel('Marks')
plt.title('Hour studies vs marks')
plt.show()
```

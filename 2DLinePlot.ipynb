```python
#if we have to compare two lines:
import pandas as pd
data = {
    "V Kohli":[100,200,300,350,900],
    "Rohit Sharma":[200,100,200,500,800],
    "year":[2020,2021,2022,2023,2024]
}
df = pd.DataFrame(data)
df
```

```python
plt.plot(df['year'],df['V Kohli'], linestyle='dotted', label='Virat')
plt.plot(df['year'],df['Rohit Sharma'], label='Rohit')
plt.xlabel("Scores")
plt.ylabel("Year")
plt.title("Rohit and Virat Scores per year")
plt.legend() #make label works
plt.grid() #add gird lines to the plot
```
